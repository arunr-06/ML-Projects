# Credit Card Fraud Detection

### Purpose: 
To help credit card companies determine the authencity of transactions - legit or fraudulent

### Dataset:
Dataset has been sourced from Kaggle & named as creditcard.csv (Please zip the file before beginning the project)

### Context: 
Dataset is a highly unbalanced one with 492 fraudelent transaction out of total 284807 transactions  
* Columns are named v1,v2 and so on for confidentiality purpose 
* Variables contained in the columns are converted to numeric values using PCA transformation
* Column Class has two values 0 and 1, with 0 denoting legit transaction & 1 denoting fraudelent transaction.

### Model:
Logistic Regression model

**Step 1:**
Importing the necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Step 2:** Importing the CSV file and storing it in a DataFrame named df

```python
df = pd.read_csv('creditcard.csv')
```

**Step 3:** Reading the first five rows of df 

```python
df.head()
```

**Step 4:** Finding details - name of all columns & data types contained in each column

```python
df.info()
```

**Step 5:** Finding the count of null values in each column

```python
df.isnull().sum()
```

**Step 6:** Finding the number of rows for each value in column Class

```python
df['Class'].value_counts()
```

**Step 7:** Dividing df into two DataFrames legit & fraud with legit containing rows with Class 0 & fraud containing rows with Class 1
***
*Data is highly unbalanced with count of values in Class 0 high compared to Class 1, hence the division*
***

```python
legit = df.loc[df['Class']==0]
fraud = df.loc[df['Class']==1]
```

**Step 8:** Finding the descriptive statistics of legit & fraud

```python
legit['Amount'].describe()
```

```python
fraud['Amount'].describe()
```

**Step 9:** Finding the average of all attributes of Class 0 & Class 1

```python
df.groupby('Class').mean()
```

**Step 10**: Creating a DataFrame legit_sample by picking 492 rows in random from legit 
***
*Undersampling to balance the unbalanced/uneven dataset by decreasing the size of Class 0. 
At the end of the process, both fraud & legit transactions will be equal in count - 492.* 
***

```python
legit_sample = legit.sample(n=492)
legit_sample.head()
```

**Step 11:** Concatenating two DataFrames - fraud & legit_sample into 1 - df1

```python
df1 = pd.concat([legit_sample, fraud])
df1
```

**Step 12:** Finding the number of rows for each value in column Class

```python
df1['Class'].value_counts()
```

**Step 13:** Finding the average of all attributes of Class 0 & Class 1 in df1 to ensure that it's distribution is similar to df 

```python
df1.groupby('Class').mean()
```

**Step 14:** Splitting data into features(X) & target(Y) 
*** 
*Features are the input, the variables fed to the machine & target is the output, the variable to be predicted.*

*Column class is the label & other columns are the features.*
***

```python
X = df1.drop('Class',axis=1)
X
```

```python
Y = df1['Class']
Y
```

**Step 15:** Splitting features X & labels Y randomly into training & test 
***
*20% test - X_test, Y_test | 80% training - X_train, Y_train*

*X_train will contain features and Y_train will contain the corresponding label*

*Stratify will ensure that a similar proportion of Class is picked in both test & training*

*random_state = 2 will ensure that the same result appear always*
***

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, 
                                                    stratify= Y, random_state = 2)
```

**Step 16:**
Display the number of rows & columns in training & test data

```python
X.shape, X_train.shape, X_test.shape
```

```python
Y.shape, Y_train.shape, Y_test.shape 
```

***
*Logistic regression model is selected as it is good for binary classification problems*
***

```python
model = LogisticRegression()
```

**Step 17:** Training the model using training data

```python
model.fit(X_train, Y_train)
```

**Step 18:** Making predictions & evaluating the model based on accuracy score
***
*X_train values will be fed to predict the values of Y_train*

*Predicted values will be compared with the original values*

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
training_data_accuracy
```

**Step 19:** Finding the `accuracy of the model in a test which the model has not seen!`

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
test_data_accuracy
```

***
*The model can predict if the `transaction is legit or fraudelent with 95% accuracy`* 

*The accuracy score is similar in both training & test data* 

*If accuracy score in training data is very high compared to test data, the model is overfitted/overtrained*

*If accuracy score in test data is very high compared to training data, the model is underfitted/undertrained*
***
